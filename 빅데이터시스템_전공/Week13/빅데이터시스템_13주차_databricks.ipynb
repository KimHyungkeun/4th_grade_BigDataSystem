```
from pyspark.sql.functions import lower, upper, month, col

homicidesNewYorkDF = (crimeDataNewYorkDF
                    .select(month(col("reportDate")).alias("month"), col("offenseDescription").alias("offense"))
                    .filter(lower(col("offenseDescription")).contains("murder") | lower(col("offenseDescription")).contains("homocide"))
                    )
display(homicidesNewYorkDF)
```

```
display(crimeDataBostonDF.select("OFFENSE_CODE_GROUP").limit(5))
```

```
homicidesBostonDF = (crimeDataBostonDF
                    .select("month", col("OFFENSE_CODE_GROUP").alias("offense"))
                    .filter(lower(col("OFFENSE_CODE_GROUP")).contains("homicide"))
                    )
display(homicidesBostonDF)
```

```
homicidesBostonAndNewYorkDF = homicidesNewYorkDF.union(homicidesBostonDF)
```

```
display(homicidesBostonAndNewYorkDF.orderBy("month").limit(5))
```

```
from pyspark.sql.functions import month, col, lower
crimeDataChicagoDF.select("primaryType").filter(lower(col("primaryType")).contains("homicide")).show(3)
```

```
homicidesChicagoDF = crimeDataChicagoDF\
                  .select(month(col("date")).alias("month"), col("primaryType").alias("offense"))\
                  .filter(lower(col("offense")).contains("homicide"))
display(homicidesChicagoDF)
```

```
allHomicidesDF = homocidesNewYorkDF.union(homicidesBostonDF).union(homicidesChicagoDF)
display(allHomicidesDF)
```

```
from pyspark.sql.functions import count
homicidesByMonthDF = allHomicidesDF.select("month").groupBy("month").count().orderBy("month")
display(homicidesByMonthDF)
```

```
homicidesByMonthBostonDF = homicidesBostonDF.select("month").groupBy("month").count().orderBy("month")
homicidesByMonthNewYorkDF = homicidesNewYorkDF.select("month").groupBy("month").count().orderBy("month")
homicidesByMonthChicagoDF = homicidesChicagoDF.select("month").groupBy("month").count().orderBy("month")
```

```
from pyspark.sql.functions import lit
homicidesByMonthChicagoDF = homicidesByMonthChicagoDF.withColumn("city", lit("Chicago"))
homicidesByMonthChicagoDF.show(3)
```

```
from pyspark.sql.functions import lit

combinedhomicidesByMonthDF = (homicidesByMonthChicagoDF.withColumn("city", lit("Chicago"))
                             .union(homicidesByMonthBostonDF.withColumn("city", lit("Boston")))
                             .union(homicidesByMonthNewYorkDF.withColumn("city", lit("New York")))
                             )
display(combinedhomicidesByMonthDF)
```
