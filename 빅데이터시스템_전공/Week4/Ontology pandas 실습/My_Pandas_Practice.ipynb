# Random Shuffling
- Pandas에 Sample 함수를 통해 shuffling
- sample() : random하게 샘플을 추출하는 함수, frac를 통해 비율을 정함
- reset_index() : index를 재설정 하는 함수


# 1. sample() 함수

```python
# 총 30개의 데이터로 이루어진 DataFrame 중 10%를 random함게 샘플을 추출하면
# 3개의 데이터로 이루어진 DataFrame이 생성

print("tripleDF length : ", len(tripleDF))
testDF = tripleDF.sample(frac=0.1)
print("testDF length : ", len(testDF))
testDF
```

```python
# frac = 1 이라는 건 모든 데이터를 추출한다는 뜻
# 모든 데이터를 Random Shuffling() 한 효과를 얻을 수 있음

randomTripleDF = tripleDF.sample(frac=1)
randomTripleDF.head()
```

# 2. reset_index() 함수 

```python
# index는 reset되었지만 사용된 index가 column으로 올라가기 때문에
# 이를 삭제하는 drop argument를 사용 (drop=True)

randomTripleDf = tripleDF.sample(frac=1).reset_index(drop=True)
randomTripleDF.head()
```

# 3. subClassOf, hasFather, ObjectProperty DataFrame 생성

- Boolean indexing으로 Filtering

# 1. Boolean indexing Example

```python
test_df = pd.DataFrame({'id' : ['a','b','c','d','e','f'], 'var':[1,2,3,4,5,6]})
test_df
```

```python
# boolean series
print("----boolean series----")
bool_series = test_df['var'] > 4
print(type(bool_series))
print()
print(bool_series)
print()

# boolean DataFrame
print("----boolean dataframe----")
bool_df = test_df[['var']] > 4
print(type(bool_df))
print()
bool_df
```

# 1-1 boolean indexing 시 dataframe를 argument로 넘겨줄 때

```python
# boolean indexing 후 원하는 값을 추출하지 못함

indexing_df1 = test_df[bool_df]
indexing_df1
```

# 1-2 boolean indexing 시 series를 argument로 넘겨줄 때

```python
# boolean indexing 후 원하는 값을 추출할 수 있음
# 따라서 boolean indexing을 할 때는 series 형태의 argument,를 넘겨야 함

indexing_df2 = test_df[bool_series]
indexing_df2
```

```python
type(indexing_df2)
```

# 1-3 여러 비교 조건문을 활용한 boolean indexing
- 여러개의 비교 조건문은 &(and), |(or)로 연결하고 조건문을 반드시 괄호()로 해야함

```python
test_df
```
