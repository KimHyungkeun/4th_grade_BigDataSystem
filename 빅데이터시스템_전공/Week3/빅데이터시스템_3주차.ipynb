```python
import numpy as np
import matplotlib.pyplot as plt
N=500
X=2*np.random.rand(500,1) #uniform distribution
y=4+ 3*X +np.random.randn(500,1) #standard normal distribution

print(X.shape)
print(y.shape)
print(max(X))
print(min(X))
print(max(y))
print(min(y))
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X,y)
plt.title("Linear Regression Data")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
theta1 = 3
theta0 = 4
theta = np.array([[theta1, theta0]]).T
print(theta)
```

```python
oneColumns = np.full((N,1),1)
print(oneColumns.shape)
X1 = np.column_stack((X,oneColumns))
print(X1.shape)
print(X1[:10,:])
```

```python
predictions = np.matmul(X1, theta)
print(predictions.shape)
plt.figure(figsize=(10,10))
plt.scatter(X,y)
plt.plot(X, predictions, color = "red")
plt.title("Linear Regression Data")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
a = np.array([[1],[2],[3]])
b = np.array([[2],[4],[1]])
c = a - b
print(a)
print(b)
print(c)
d = np.square(c)
print(d)
print(np.sum(d))

print(X1.shape)
print(X1.shape[0])
print(y.shape)
print(predictions.shape)

def cost(X,y,theta) :
    m = X.shape[0]
    predictions = np.matmul(X, theta)
    diff = predictions - y
    cost = (1/2*m)*np.sum(np.square(diff))
    return cost

print(cost(X1,y,theta))
```

```python
def linear_regression(X, theta, y, alpha) :
    m = X.shape[0]
    num_iterations = 100
    costs = []
    thetas = [theta]
    for i in range(num_iterations) :
#         print("predictions : " + str(predictions))
#         print("y : " + str(y))
        predictions = np.matmul(X1, theta)
        errors = predictions - y
        delta = np.matmul(X.T, errors)
        
        theta = theta - (alpha/m)*delta
        cost1 = cost(X,y,theta)
        costs.append(cost1)
        thetas.append(theta)
    return costs, thetas

costs, thetas = linear_regression(X1, theta, y, 0.5)
print(len(costs), len(thetas))
plt.plot(costs)
```

```python
initialTheta = thetas[0]
print(initialTheta)
finalTheta = thetas[-1]
print(finalTheta)
prediction1 = np.matmul(X1, initialTheta)
prediction2 = np.matmul(X1, finalTheta)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X,y)
plt.plot(X, prediction1, color = "red")
plt.plot(X, prediction2, color = "blue")
plt.title("Linear Regression Data")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
